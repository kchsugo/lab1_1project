```python
#2번

import matplotlib.pyplot as plt
from matplotlib_venn import venn3_unweighted

diction={'a': [0, 1, 2, 3], 'b': [0, 2, 4, 6], 'c': [1, 2, 5, 7]}
split=list(diction.values())

for i in range(len(split)):
    for j in range(len(split[0])):
        split[i][j]= str(split[i][j])

A=set(split[0])
B=set(split[1])
C=set(split[2])
v = venn3_unweighted([A,B,C], set_labels=('A','B','C'))




if(v.get_label_by_id('100')):
    v.get_label_by_id('100').set_text('\n'.join(A-B-C))
if(v.get_label_by_id('010')):
    v.get_label_by_id('010').set_text('\n'.join(B-C-A))
if(v.get_label_by_id('001')):
    v.get_label_by_id('001').set_text('\n'.join(C-A-B))
if(v.get_label_by_id('110')):
    v.get_label_by_id('110').set_text('\n'.join(A&B-C))
if(v.get_label_by_id('101')):
    v.get_label_by_id('101').set_text('\n'.join(A&C-B))
if(v.get_label_by_id('011')):
    v.get_label_by_id('011').set_text('\n'.join(B&C-A))
if(v.get_label_by_id('111')):
    v.get_label_by_id('111').set_text('\n'.join(B&C&A))
    

plt.title("venn diagram")
plt.show()
```
